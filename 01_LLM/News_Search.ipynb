```python
from langchain_groq import ChatGroq
from langchain.agents import create_agent
from langchain.tools import tool
from langchain_community.utilities import SerpAPIWrapper


llm = ChatGroq(
    model="qwen/qwen3-32b",
    temperature=0,
    streaming=True
)

search = SerpAPIWrapper(serpapi_api_key="d1261e98e4086a724960d48decefe0d1529195a3761899f1e7f510c50b8e1a22")

```

```python

@tool
def google_search(query: str) -> str:
    """Search Google for latest news and info."""
    return search.run(query)

agent = create_agent(
    model=llm,
    tools=[google_search],
    system_prompt="You are a news assistant. Use google_search for latest info."
)

messages = {
        "messages": [
            {"role": "user", "content": "What is the latest news about ISRO?"}
        ]
    }
for chunk in agent.stream(messages,stream_mode="values"):
    print(chunk["messages"][-1].content, end="", flush=True)


```
